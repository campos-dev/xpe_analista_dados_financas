```python
product_value_fixed = df['Valor da venda'] / df['Quantidade unitária']

df['Valor do produto'] = df['Valor do produto'].fillna(value=product_value_fixed)
```

```python
sell_value_fixed = df['Quantidade unitária'] * df['Valor do produto']

df['Valor da venda'] = df['Valor da venda'].fillna(value=sell_value_fixed)
```

```python
# 'Valor da venda' mean from Minas Gerais
# Qual é a média do valor das vendas de todos os produtos destinados ao Estado de Minas Gerais (MG)?
products_MG = df[df['Localização']=='MG']
products_MG['Valor da venda'].mean()
```

```python
# best-selling product
# Qual produto possui o maior número de vendas? 
df['Produto vendido'].value_counts()
```

```python
# Standard Deviation of 'Quantidade unitária' from Pernambuco
# Qual é o desvio padrão da quantidade unitária para todos os produtos vendidos no Estado de Pernambuco (PE)?
products_PE = df[df['Localização'] == 'PE']
products_PE['Quantidade unitária'].std()
```

```python
# Max of "Valor da venda" from Santa Catarina
# Qual é o valor máximo de venda registrado para os produtos comercializados no Estado de Santa Catarina (SC)?
products_SC = df[df['Localização'] == 'SC']
products_SC['Valor da venda'].max()
```

```python
# Sum of 'Valor da venda' from 'Produto A'
# Somatório do valor total das vendas de todos os produtos identificados como "Produto A". 
product_a = df[df['Produto vendido'] == 'Produto A']
product_a['Valor da venda'].sum()
```

```python
# Which product has more 'Quantidade unitária'
# qual produto teve o maior total de unidades vendidas?
df.groupby('Produto vendido')['Quantidade unitária'].sum()
```

```python
# Mean of 'Quantidade unitária' of 'Produto B' in Pernambuco
# Qual é a média da quantidade unitária vendida para o Produto B no Estado de Pernambuco (PE)?
product_B = df[df['Produto vendido'] == 'Produto B']
product_B_PE = product_B[product_B['Localização'] == 'PE']
round(product_B_PE['Quantidade unitária'].mean(),2)
```

```python
# Sum of 'Valor da venda' in MG and SP
# Qual é a soma do valor da venda para todos os produtos vendidos para os estados de MG e SP?

products_SP_MG = df[df['Localização'].isin(['SP', 'MG'])]
products_SP_MG['Valor da venda'].sum()
```

```python
# Mean of Valor do produto" from AL, BA, CE, MA, PB, PE, PI, RN, SE
# média da coluna "Valor do produto" para todos os produtos que estão disponíveis nos estados de AL, BA, CE, MA, PB, PE, PI, RN, SE

products_NE = df[df['Localização'].isin(['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'])]
round(products_NE['Valor do produto'].mean(),2) 
```
